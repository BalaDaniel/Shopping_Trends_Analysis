```python
import pandas as pd
```

```python
shoppingdf=pd.read_csv('E:/Learning/Edureka/Projects/shopping_trends.csv')
shoppingdf
```

### Geder wise Sales

```python
purchases_genderwise=shoppingdf.groupby('Gender')['Purchase Amount (USD)'].sum().reset_index()
purchases_genderwise
```

```python
total_purchases=shoppingdf['Purchase Amount (USD)'].sum()
total_purchases
```

```python
purchases_genderwise['Percentage']=(purchases_genderwise['Purchase Amount (USD)']/total_purchases*100).round(2)
purchases_genderwise
```

```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure(figsize=(8,8))
plt.pie(purchases_genderwise['Percentage'],labels=purchases_genderwise['Gender'],autopct='%1.1f%%',startangle=180)
plt.title('Purchase Percentage between male and female')
fig.subplots_adjust(left=1, right=4, top=0.7, bottom=0.3)
plt.show()
```

### Category wise total purchase amount

```python
purchases_categorywise=shoppingdf.groupby('Category')['Purchase Amount (USD)'].sum().reset_index()
purchases_categorywise
```

```python
purchases_categorywise['Percentge']=(purchases_categorywise['Purchase Amount (USD)']/total_purchases*100).round(2)
purchases_categorywise_sort=purchases_categorywise.sort_values(by=['Purchase Amount (USD)'],ascending=False)
purchases_categorywise_sort
```

```python
plt.figure(figsize=(6,6))
plt.pie(purchases_categorywise_sort['Percentge'],labels=purchases_categorywise_sort['Category'],autopct='%1.2f%%',startangle=140)
plt.title('Percent of purchases categorywise')
plt.show()
```

### Creating Age_Group column

```python
bins = [18, 30, 50, 60, 100]
labels = ['18-29', '30-49', '50-59', '60-100']
shoppingdf['Age_group'] = pd.cut(shoppingdf['Age'], bins=bins, labels=labels, right=False)
shoppingdf
```

### Total Purchased amount by Age_Group

```python
purchases_by_agegroup=shoppingdf.groupby('Age_group')['Purchase Amount (USD)'].sum().reset_index().sort_values(by='Purchase Amount (USD)',ascending=False)
purchases_by_agegroup
```

```python
plt.figure(figsize=(5,5))
plt.bar(purchases_by_agegroup['Age_group'],purchases_by_agegroup['Purchase Amount (USD)'])
for i, value in enumerate(purchases_by_agegroup['Purchase Amount (USD)']):
    plt.text(i, value, str(value), ha='center', va='bottom')
plt.xlabel('Age_group')
plt.ylabel('Purchase Amount (USD)')
plt.title('Total Purchased amount by Age Group')
plt.show()
```
